```python
import numpy as np
import pandas as pd
from scipy.stats import norm
```

```python
df = pd.read_csv('advertisement_clicks.csv')
```

```python
df.head()
```

```python
xA = df[df['advertisement_id']=='A']['action'].to_numpy()
xB = df[df['advertisement_id']=='B']['action'].to_numpy()

nA = len(xA)
nB = len(xB)

pA_hat = xA.mean()
pB_hat = xB.mean()

p_pool = (xA.sum() + xB.sum()) / (nA + nB)
s_hat = np.sqrt(p_pool * (1 - p_pool) * (1/nA + 1/nB))

z = (pA_hat - pB_hat) / s_hat
p = 2 * (1 - norm.cdf(abs(z)))

z, p
```

The Z-test for the difference in click-through rates between Advertisement A and Advertisement B yielded a Z-value of approximately -3.21 and a p-value of 0.0013. Since the p-value is much smaller than 0.05, we reject the null hypothesis. This indicates that there is a statistically significant difference in click-through rates between the two advertisements, with Advertisement B performing better than Advertisement A.
